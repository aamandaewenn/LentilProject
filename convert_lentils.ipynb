```python
import numpy as np
import matplotlib.pyplot as plt
import cv2
import glob
from skimage import io
import os
from skimage import data, color
import numpy as np
import matplotlib.pyplot as plt
```

```python
# read in all the paths from the specified folder
image_paths = sorted(glob.glob('/Users/are390/Documents/project/Testa/**/**/*.npy'))
# for each path, load the image and convert it to correct type
for item in image_paths: 
    image = io.imread(item)
    plt.imshow(image)
    plt.show()


    # create a path to save the image to
    out_dir = os.path.join('../../Documents/project/Testa/Converted', *item.split('/')[-3:-1])
    os.makedirs(out_dir, exist_ok=True)
    out_name = item.split('/')[-1][:-4]+'.png'
    
    # save the image
    io.imsave(os.path.join(out_dir, out_name), image)
```
